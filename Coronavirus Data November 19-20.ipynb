```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup 
```

```python
states = ["Alabama", "Alaska", "Arizona", "Arkansas", "California", "Colorado", "Connecticut", "Delaware", "Florida", 
          "Georgia", "Hawaii","Idaho","Illinois", "Indiana","Iowa","Kansas","Kentucky","Louisiana","Maine","Maryland",
          "Massachusetts","Michigan","Minnesota", "Mississippi","Missouri","Montana","Nebraska","Nevada", "New Hampshire",
          "New Jersey","New Mexico","New York","North Carolina","North Dakota","Ohio","Oklahoma","Oregon","Pennsylvania",
          "Rhode Island","South Carolina","South Dakota","Tennessee","Texas","Utah","Vermont","Virginia","Washington",
          "West Virginia","Wisconsin","Wyoming"]
statesnew = []
```

```python
for s in states:
    s = s.replace(" ", "-")
    s = s.lower()
    statesnew.append(s)
```

```python
link = "https://usafacts.org/visualizations/coronavirus-covid-19-spread-map/state/arizona"
```

```python
#New Final Tester
listofdicts = []
for state in statesnew:
    URL = "https://usafacts.org/visualizations/coronavirus-covid-19-spread-map/state/" + state
    page = requests.get(URL, headers={'User-Agent': 'Mozilla/5.0'})
    soup = BeautifulSoup(page.content, 'html.parser')
    table = soup.find_all('table') 
    rows = table[0].find_all('tr', class_="MuiTableRow-root-127")
    for i in range (1, len(rows)):
        dict = {}
        dict['State Name'] = state
        rownumber = rows[i]
        firstcolumn = rownumber.find_all('th')
        dict['County Name'] = firstcolumn[0].text
        othercolumns = rownumber.find_all('td')
        dict['Known Number of Cases'] = othercolumns[0].text
        dict['Deaths'] = othercolumns[1].text
        dict['Known Number of Cases per 100k'] = othercolumns[2].text
        listofdicts.append(dict)
        
```

```python
dataframe = pd.DataFrame(listofdicts)  
    
# Print the data  
dataframe
```

```python
dataframe.to_csv('Covid.csv')
```
