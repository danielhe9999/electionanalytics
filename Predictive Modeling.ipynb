```python
%matplotlib inline
#imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression

#reading in UniversalBank csv and setting it as a Data Frame called df
df = pd.read_csv('completedataset.csv')
```

```python
import pandas as pd
from sklearn import linear_model
import statsmodels.api as sm
```

```python
df = df.set_index('County + State')
```

```python
df
```

```python
df.dtypes
```

```python
voters = df['Total number of voters'].tolist()
income = df['Estimate Median Household Income'].tolist()
cases = df['Known Number of Cases per 100k'].tolist()
```

```python
newvoters = []
newincome = []
newcases = []
for item in voters:
    try:
        newstr = item.replace(',','')
        newstr = int(newstr)
        newvoters.append(newstr)
    except:
        newstr = item
        newstr = int(newstr)
        newvoters.append(newstr)
```

```python
for item in income:
    try:
        newstr = item.replace(',','')
        newstr = float(newstr)
        newincome.append(newstr)
    except:
        newstr = item
        newstr = float(newstr)
        newincome.append(newstr)
```

```python
for item in cases:
    try:
        newstr = item.replace(',','')
        newstr = float(newstr)
        newcases.append(newstr)
    except:
        newstr = item
        newstr = float(newstr)
        newcases.append(newstr)
```

```python
df['Total number of voters'] = newvoters
df['Median Income'] = newincome
df['Cases per 100k'] = newcases
```

```python
df = df.drop(['Total number of voters', 'Estimate Median Household Income', 'Known Number of Cases per 100k'], axis = 1)
```

```python
df
```

```python
#mulitple regression 

X = df.drop(['Result'], axis = 1)
Y = df['Result']
```

```python

regr = linear_model.LinearRegression()
regr.fit(X, Y)

print('Intercept: \n', regr.intercept_)
print('Coefficients: \n', regr.coef_)


```

```python
# with statsmodels
X = sm.add_constant(X) # adding a constant
 
model = sm.OLS(Y, X).fit()
predictions = model.predict(X) 
 
print_model = model.summary()
print(print_model)
```
