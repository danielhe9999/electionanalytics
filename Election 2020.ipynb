```python
import bs4 
from selenium import webdriver 
driver = webdriver.Chrome(r"C:\Users\danie\Independent Data Science Projects\chromedriver")
listofdicts = []

for link in listoflinks:
    nameofstate = link.split("-")[1]
    try:
        driver.get(link)
        html = driver.page_source
        soup = bs4.BeautifulSoup(html,'html.parser')  
        containers = soup.find_all('table', {'class': 'e-table e-county-table'})
        tablerow = containers[0].find_all('tr', {'class': 'e-row'})
        for i in range (0, len(tablerow)-1): #for each county
            dict = {}
            dict["Name of State"] = nameofstate
            row = tablerow[i]
            countyname = row.find_all('th', {'class': 'e-cell e-name'})
            name = countyname[0].text
            dict["County Name"] = name
            margin = row.find_all('td', {'class': 'e-cell e-margin'})
            marginamount = margin[0].text
            dict["Margin Amount"] = marginamount
            totalvoters = row.find_all('td', {'class': 'e-cell e-total e-v'})
            votes = totalvoters[0].text
            dict["Total number of voters"] = votes
            listofdicts.append(dict)
    except:
        print("didn't work")
```

```python

dataframe = pd.DataFrame(listofdicts)  
    
# Print the data  
dataframe.tail(10)
```
